```python
import pandas as pd
```

Preperation of the input data:
- Reading in all JSON input files
- adding 1000000*gameNumber to the index to include the game-ID in the tickCOunt
- make tickCount the index
- combine all dataframes


```python
def prepareInputDf():
  input_df = pd.DataFrame(columns=['buttons', 'mouseDeltaX', 'mouseDeltaY'])
  path = 'drive/MyDrive/BA_JSON/'
  files = ['cs02_input.json', 'cs03_input.json', 'cs04_input.json', 
           'cs06_input.json', 'cs07_input.json', 'cs08_input.json', 
           'cs09_input.json', 'cs10_input.json', 'cs12_input.json']
           # 'cs01_input.json', 'cs11_input.json', 'cs13_input.json'
           # These three are missing due to wrong mouse input data
  game_index_factor = 1000000
  game_index = 0
  
  for x in files:
    file = path + x
    df = pd.read_json(file)
    df['tickCount'] += game_index_factor*game_index
    game_index += 1
    df = df.set_index('tickCount')
    df = df[~df.index.duplicated(keep='first')]
    input_df = pd.concat([input_df, df])
  return input_df
```

```python
input_df = prepareInputDf()
```
